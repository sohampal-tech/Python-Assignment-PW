2. What is an interpreter in Python ?
  - An interpreter in Python reads and executes code line by line. It translates Python code into machine-readable instructions during runtime.

3. What are pre-defined keywords in Python?
  - Keywords are reserved words that have special meaning in Python. Examples:  
  if, else, while, for, def, class, return.
  They cannot be used as variable names.

4. Can keywords be used as variable names?
  - No, keywords cannot be used as variable names because they are reserved by Python for specific functionality.

5. What is mutability in Python?
  - Mutability refers to whether an object can be changed after it is created.
  .Mutable: Can be changed (e.g., list, dict)
  .Immutable: Cannot be changed (e.g., tuple, string)

6. Why are lists mutable, but tuples are immutable?
  - Lists are designed to hold dynamic data and allow changes (like append, remove).
 .Tuples are meant to be fixed collections, making them faster and safe from unintended changes.

7. What is the difference between “==” and “is” operators in Python?
  - ==: Compares values (e.g., 5 == 5 → True)
  
  is: Compares memory locations (whether two variables point to the same object)

8. What are logical operators in Python?
  - Logical operators are used to combine conditional statements:
  
  .and: True if both conditions are true
  
  .or: True if at least one condition is true
  
  .not: Inverts the result

9. What is type casting in Python?
   - Type casting in Python is the process of converting the value of one data type to another.It is done using built-in functions like int(), float(), str(), etc.

10. What is the difference between implicit and explicit type casting?
    - Implicit: Automatically done by Python (e.g., int + float → float)

  Explicit: Manually done using functions like int(), str(), float()

11. What is the purpose of conditional statements in Python?
   - Conditional statements allow decision-making in code. Based on conditions, different blocks of code are executed (e.g., if, elif, else).

12. How does the elif statement work?
   - elif stands for "else if". It checks another condition if the previous if was False. Only the first true condition block is executed.

13. What is the difference between for and while loops?
    - for: Iterates over a sequence (e.g., list, range)- when the number of iterations is known.

  while: Repeats as long as a condition is true - when the number of iterations is unknown.

14. Describe a scenario where a while loop is more suitable than a for loop?
   - When you need to loop until a condition is met but don't know how many times in advance.Example: Keep asking the user for a password until they enter the correct one.
