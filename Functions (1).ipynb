## **6. Explain the concept of generators in Python and how they are defined.**

 Answer :

-  A generator is a function that returns values one at a time using the yield keyword.

-  Example:

```python
def count_up():
    yield 5
    yield 4

for i in count_up():
    print(i)
```

## **7. What are the advantages of using generators over regular functions?**
  
 Answer:

- Saves memory (values generated one at a time)

- Good for large data or infinite sequences

- Faster performance in some cases
-  Example :

```python
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1

# Using the generator
for num in count_up_to(5):
    print(num)
```

## **9. Explain the purpose and usage of the map() function in Python.**

 Answer:

-  map() applies a function to each item in an iterable like a list.

-  Example:

```python
nums = [1, 2, 3]
doubled = list(map(lambda x: x * 2, nums))
print(doubled)  # Output: [2, 4, 6]
```

## **10. What is the difference between map(), reduce(), and filter() functions in Python?**

Answer:

-  map() applies a function to all items.

-  filter() keeps items that match a condition.

-  reduce() (from functools) reduces a list to a single value.

-  Example:

```python
from functools import reduce
nums = [1, 2, 3, 4]
print(list(map(lambda x: x*2, nums)))       # [2, 4, 6, 8]
print(list(filter(lambda x: x % 2 == 0, nums)))  # [2, 4]
print(reduce(lambda x, y: x + y, nums))     # 10
```

## **1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.**

```python
def sum_even(lst):
    return sum(num for num in lst if num % 2 == 0)

print(sum_even([1, 2, 3, 4, 5]))
```

## **4. Write a Python function that checks if a given number is prime or not from 1 to 200.**

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

for i in range(1, 201):
    if is_prime(i):
        print(i, end=" ")
```

## **5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.**

```python
class Fibonacci:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        val = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return val

for num in Fibonacci(10):
    print(num, end=" ")
```

## **6. Write a generator function in Python that yields the powers of 2 up to a given exponent.**

```python
def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i

for val in powers_of_two(5):
    print(val)
```

## **7. Implement a generator function that reads a file line by line and yields each line as a string.**

```python
# Create a sample file to test
with open("example.txt", "w") as f:
    f.write("Hello PW\n")
    f.write("good morning\n")
    f.write("This is a test file\n")
```

```python
def read_file_lines(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()
```

```python
for line in read_file_lines("example.txt"):
    print(line)
```

## **10. Create a Python program that uses filter() to remove all the vowels from a given string.**

```python
def remove_vowels(text):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda x: x not in vowels, text))

print(remove_vowels("Hello World"))
```
